```python
import pandas as pd
```

```python
data903 = pd.read_csv('TrafficPtrS903_2007-01_2018-10_All.csv/TrafficVolumeByHour_S903_2007-01_2018-10.csv')
```

```python
data903.rename(columns={'TravelDirection':'direction', 'Date':'date', 'DayOfWeek':'day'}, inplace=True)
```

```python
data903.T
```

```python
date = pd.to_datetime(data903['date'])
```

```python
date.head()
```

```python
data903.direction[data903.direction == 'Eastbound'] = 'EB'
```

```python
data903.direction[data903.direction == 'Westbound'] = 'WB'
```

```python
data903.T
```

```python
westbound = pd.Series([True if value == 'WB' else False for value in data903.direction])
```

```python
westbound
```

```python
westbound.index = date.index
```

```python
data = data903.drop(['SiteId', 'SiteLocation', 'date', 'direction'], axis=1)
```

```python
data.T
```

```python
df = pd.concat([date, westbound, data], axis=1)
```

```python
df.T
```

```python
df.rename(columns={0:'westbound'}, inplace=True)
```

```python
df
```

```python
cleaned_df = df[(df['date'] > '2017-12-31') & (df['date'] < '2018-04-03')]
```
