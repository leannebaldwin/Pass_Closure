```python
def f(row):
    if row['sknt'] >= 10.00:
        val = True
    elif row['gust'] >= 20.00:
        val = True
    else:
        val = False
    return val
```

```python
wind_df = pd.concat([wind_speed, gust_speed], axis=1)
```

```python
windy = wind_df.apply(f, axis=1)
```
